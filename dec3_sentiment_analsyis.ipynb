```python
import datetime, requests, yfinance
from alpaca_trade_api.rest import REST
from getpass import getpass
BASE_URL="https://paper-api.alpaca.markets"
API_KEY= getpass('Enter your API KEY: ')
SECRET_KEY= getpass('enter your SECRET KEY:')
api = REST(key_id= API_KEY, secret_key= SECRET_KEY, base_url= BASE_URL)




```

```python
news= api.get_news('TSLA')
news
```

```python
news= api.get_news('TSLA', start='2023-05-19', end='2023-10-09', include_content=True)
news
```

```python
##to fetch news articles and images
for story in news:
  print(story.headline)
  print(story.summary)
  for image in story.images:
    print(image['url'])
```
